```python
import folium
from folium.plugins import HeatMap

#Data contained Nan values so it needs to be removed
df = df.dropna(subset=['Start_Lat', 'Start_Lng'])

coordinates = list(zip(df['Start_Lat'], df['Start_Lng']))

#mean helps in taking average of both Lat and Lng
m = folium.Map(location=[df['Start_Lat'].mean(), df['Start_Lng'].mean()], zoom_start=10)

HeatMap(coordinates).add_to(m)

```

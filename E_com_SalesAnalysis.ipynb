```python
import pandas as pd
import os
```

**Combining all csv files in a single file for better analysis**

```python
combined_data = pd.DataFrame()

# List of CSV files to concatenate
csv_files = ["/content/Sales_January_2019.csv","/content/Sales_February_2019.csv", "/content/Sales_March_2019.csv","/content/Sales_April_2019.csv","/content/Sales_May_2019.csv","/content/Sales_June_2019.csv","/content/Sales_July_2019.csv","/content/Sales_August_2019.csv", "/content/Sales_September_2019.csv","/content/Sales_October_2019.csv","/content/Sales_November_2019.csv","/content/Sales_December_2019.csv"]

# Step 3: Loop through the CSV files
for file in csv_files:
    # Step 4: Read each file into a DataFrame
    data = pd.read_csv(file)

    # Step 5: Append the data to the combined DataFrame
    combined_data = combined_data.append(data, ignore_index=True)

# Specify the output file where you want to save the combined data
output_file = "combined_data.csv"

# Write the combined data to a new CSV file
combined_data.to_csv(output_file, index=False)

print(f"Combined data saved to {output_file}")
```

```python
combined_data
```

**Data Cleaning**

```python
#Indentification of Nan values
nan_df=combined_data[combined_data.isna().any(axis=1)]
nan_df
```

```python
combined_data=combined_data.dropna(how='all')
combined_data
```

```python
#CSV file contains column name repeated in multiple rows
temp=combined_data[combined_data["Order ID"].str[0:2]=='Or']
temp

# There are 355 rows containing duplicates with column names
```

```python
combined_data=combined_data[combined_data["Order ID"].str[0:2]!='Or']        #Duplicate values are hidden
```

Adding a month column

```python
#Creating a separate column for each month for calculation purpose

combined_data["Month"]=combined_data["Order Date"].str[0:2]
combined_data["Month"]=combined_data["Month"].astype('int32')
combined_data
```

```python
combined_data.dtypes
```

**Finding out sales of each month**

```python
#Some data in either of the columns contains string value
#to_numeric converts to int or float depending on the supplied data

combined_data["Quantity Ordered"]=pd.to_numeric(combined_data["Quantity Ordered"])
combined_data["Price Each"]=pd.to_numeric(combined_data["Price Each"])
combined_data
```

```python
combined_data["Sales"]=combined_data["Quantity Ordered"]*combined_data["Price Each"]
combined_data
```

**Finding out best month of total sales**

```python
results=combined_data.groupby("Month").sum()
results
```

```python
import matplotlib.pyplot as plt

months=range(1,13)
plt.bar(months,results["Sales"])
plt.xticks(months)
plt.ylabel("Sales in million $")
plt.xlabel('Month number')
plt.show()
```

**To find sales from each city**

```python
#Address column needs to break down into city column

city_state_pattern = r',\s(.*?),\s([A-Z]{2})'

extracted_data = combined_data['Purchase Address'].str.extract(city_state_pattern)
combined_data['City'] = extracted_data[0]
combined_data['State'] = extracted_data[1]

#Separate column has been created for city
```

```python
results1=combined_data.groupby("City").sum()
results1
```

```python
results2=combined_data.groupby("State").sum()
results2
```

```python
##cities=combined_data["City"].unique()
##there are difference in results so

cities=[city for city, df in combined_data.groupby("City")]

plt.bar(cities, results1["Sales"])
plt.xticks(cities,size=8,rotation="vertical")
plt.ylabel("Sales in million $")
plt.xlabel('City name')
plt.show()
```

**What products are sold together more often**

```python
combined_data["Product"].unique()
```

```python
#Finding out orders which are placed under same order id

df=combined_data[combined_data["Order ID"].duplicated(keep=False)]
df["Grouped"]=df.groupby('Order ID')["Product"].transform(lambda x:','.join(x))

df=df[['Order ID','Grouped']].drop_duplicates()

df
```

```python
from collections import Counter             #To find two or more products which are ordered together
from itertools import combinations

count=Counter()

for row in df['Grouped']:
  row_list=row.split(',')
  count.update(Counter(combinations(row_list,2)))

##print(count)

count.most_common(15)
```

```python
#Which product has been sold the most

prod=combined_data.groupby("Product")

prod.sum()
```

```python
quantity_ordered=prod.sum()["Quantity Ordered"]

products=[product for product, df in prod]

plt.bar(products,quantity_ordered)
plt.xticks(products, rotation="vertical", size=8)
plt.ylabel("Qunatity ordered")
plt.xlabel("Product name")
plt.show()
```
